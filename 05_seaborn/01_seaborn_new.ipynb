```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.get_dataset_names()
```

```python
flights = sns.load_dataset("flights")
#flights.head(10)

flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
#flights_pivot

sns.heatmap(
    flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)
plt.title("Passengers Heatmaps")
plt.tight_layout()
```

```python
tips = sns.load_dataset("tips")

fig, ax = plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    ax=ax,
    marker="o",
    #ci=None->errorbar=None
    errorbar=None
)

ax.set_title("AVG Bill Value")
ax.set_xlabel("Days")
ax.set_ylabel("Total Bill Value")
fig.tight_layout()
```
