```python
#Normal Distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 70
sigma = 10

x=np.linspace(30, 110, 1000)
y=norm.pdf(x, mu, sigma)

#Bell Curve
plt.plot(x, y, color="black", linewidth=2, label="Normal Distribution")

x1 = np.linspace(mu - sigma, mu + sigma, 1000)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color="green", alpha=0.3, label="68% of region (1sigma)")

x2 = np.linspace(mu - 2*sigma, mu + 2*sigma, 1000)
plt.fill_between(x2, norm.pdf(x2, mu, sigma), color="yellow", alpha=0.3, label="95% of region (2sigma)")

x3 = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
plt.fill_between(x3, norm.pdf(x3, mu, sigma), color="red", alpha=0.3, label="99.7% of region (3sigma)")

plt.title("Normal Distribution of Student Scores")
plt.xlabel("Scores")
plt.ylabel("Probability Density")
plt.legend()
plt.tight_layout()    
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 100_000)
y = np.sqrt(x)

plt.plot(x, y, label="y = root(x)")
plt.title("Graph of y = root(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y, label="y=sin(x)", color="blue")

plt.axvline(0, color="black", linewidth=1.5, linestyle="--", label="x=0")
plt.axhline(0, color="black", linewidth=1.5, linestyle="--", label="y=0")

plt.title("Graph of y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.show()
```
